- exploring the red wine.csv of the Wine Quality dataset in UCI Machine Learning :
- https://archive.ics.uci.edu/dataset/186/wine+quality

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/joycechung/Downloads/wine+quality/winequality-red.csv', low_memory=False, sep=';')
```

```python
len(df)
```

```python
df
```

```python
#checking the correlation between pH and citric acid levels 

df.sort_values(['pH']).plot(x= 'pH', y='citric acid', kind = 'bar')
plt.xlabel("pH")
plt.ylabel("Citric Acid levels")

plt.xticks([]) #hid the labels on x-axis because it was too crowded, 
#and I didn't want to remove the bins. 
plt.show()

#.sort_values sorts values in ascending order by default
```

```python
df.sort_values(['pH']).plot(y='pH', kind = 'bar');
#.sort_values sorts values in ascending order by default

plt.xticks([]) #hid the labels on x-axis because it was too crowded, 
#and I didn't want to remove the bins. 

plt.xlabel("Wines (sorted from lowest to highest pH)")
plt.ylabel("pH levels")
```

```python
#finding the top 10 rows where residual sugar level is highest 

# Sort the dataframe by 'residual sugar' column in ascending order
sorted_df = df.sort_values('residual sugar', ascending=False)
sorted_df.head(10)
```

```python
#checking the row where the quality is highest 
sorted_df = df.sort_values('quality', ascending=False)
sorted_df.head(1)
```

```python
#checking the row where the quality is lowest
sorted_df = df.sort_values('quality', ascending=True)
sorted_df.head(1)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['volatile acidity'], df['fixed acidity'])
#scatterplot was used since we are comparing two continuous variables.

# Set labels and title for the plot
plt.xlabel('Volatile acidity')
plt.ylabel('Fixed acidity')
plt.title('Fixed vs Volatile Acidity')

# Display the plot
plt.show()
```
