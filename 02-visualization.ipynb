This dataset attributes first names to genders, giving counts and probabilities. It combines open-source government data from the US, UK, Canada, and Australia.

Additional Information

This dataset combines raw counts for first/given names of male and female babies in those time periods, and then calculates a probability for a name given the aggregate count.  Source datasets are from government authorities:
-US: Baby Names from Social Security Card Applications - National Data, 1880 to 2019
-UK:  Baby names in England and Wales Statistical bulletins, 2011 to 2018
-Canada: British Columbia 100 Years of Popular Baby names, 1918 to 2018
-Australia:  Popular Baby Names, Attorney-General's Department, 1944 to 2019

Creating a piechart plotting the top 10 male names

```python
male = df[df["Gender"] == 'M'] #two 'df' is needed to turn male into a dataframe
male
```

```python
top_10 = male["Name"].value_counts().head(10) #getting the 'name' column from the male names dataframe
top_10 #getting the top 10 names
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot the pie chart
plt.pie(top_10.values, labels=top_10.index, autopct='%1.0f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Top 10 Male Names')
plt.show()
```

Creating a piechart plotting the top 10 female names

```python
import pandas as pd
import matplotlib.pyplot as plt

female = df[df["Gender"] == 'F'] #two 'df' is needed to turn male into a dataframe

top_10_f = female["Name"].value_counts().head(10) #getting the 'name' column from the male names dataframe
top_10_f #getting the top 10 names

# Plot the pie chart
plt.pie(top_10_f.values, labels=top_10_f.index, autopct='%1.0f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Top 10 Female Names')
plt.show()
```

Looking at the names with the top 10 counts - what percentage of those are female, what percentage of those are male

```python
top_10_all = df["Name"].value_counts().head(10) #getting the 'name' column from all genders
top_10_all #getting the top 10 names
```

```python
# Calculate the percentage of each gender in the top 10 names
gender_percentages = df[df["Name"].isin(top_10_all.index)]['Gender'].value_counts(normalize=True) * 100

# Plot the pie chart
plt.pie(gender_percentages, labels=gender_percentages.index, autopct='%1.0f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Percentage of Male and Female in Top 10 Names')
plt.show()
```

Looking at the 26 most common names, the name most commonly starting with each letter in the alphabet A-Z 

```python
import pandas as pd

# Get the 26 most common names
top_26_names = df['Name'].value_counts().head(26).index.tolist()

# Create an empty dictionary to store the most common names for each letter
most_common_names = {}

# Iterate over each letter of the alphabet
for letter in 'ABCDEFGHIJKLMNOPQRSTUVWXYZ':
    # Filter the names starting with the current letter
    names_starting_with_letter = [name for name in top_26_names if name.startswith(letter)]
    
    # If there are names starting with the current letter
    if names_starting_with_letter:
        # Get the most common name starting with the current letter
        most_common_name = max(names_starting_with_letter, key=lambda x: df['Name'].value_counts()[x])
        
        # Store the most common name for the current letter in the dictionary
        most_common_names[letter] = most_common_name

# Print the most common name for each letter of the alphabet
for letter, name in most_common_names.items():
    print(f"The most common name starting with '{letter}' is '{name}'")
```
